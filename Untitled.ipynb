```python
from sklearn.cluster import KMeans
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
```

```python
image_path = r"C:\Users\Dosya\Pictures\Screenshots\Снимок экрана 2025-11-16 222136.png"
```

```python
image = Image.open(image_path).convert('RGB')
```

```python
image = np.array(image)
X = image.reshape(-1, 3)
plt.imshow(image)
plt.axis('off')
plt.show()
```

```python
randoming = 312
km = KMeans(n_clusters = 5, random_state = randoming)
km.fit(X)

```

```python
segmented = np.zeros_like(X)  
for i in range(5):
    segmented[km.labels_ == i] = km.cluster_centers_[i]

segmented = segmented.reshape(image.shape).astype(np.uint8)

plt.imshow(segmented)
plt.axis('off')
plt.show()

```
